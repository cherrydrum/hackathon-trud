```python
import pandas
import subscriptions
import tokenizer
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Указываем минимальное количество упоминаний слова для отображения в диаграмме
MENTIONS = 5
```

```python
# Здесь у нас процесс токенизации
data = pandas.read_csv('parsed.csv', delimiter=',')
test = []
for index, row in data.iterrows():
    variant = str(row[0] + ' ' + row[1] + ' ' + row[2])
    temp = tokenizer.tokenize(variant)
    test += list(temp)
    
print(test)
```

```python
howmuch = pandas.Series(test)
howmuch = howmuch.value_counts()

#Минимальное количество упоминаний для попадания в выборку.
howmuch = howmuch[howmuch >= MENTIONS]

#Максимальное количество упоминаний для попадания в выборку.
howmuch = howmuch[howmuch <= 2]

print(howmuch)
```

```python
x = np.arange(1, len(howmuch)+1)
y = howmuch.tolist()

fig, ax = plt.subplots()
plt.rcParams['figure.figsize'] = [20, 10]
plt.xticks(np.arange(min(x), max(x)+1, 1.0), rotation=90)
ax.plot(x, y)
ax.set_xticklabels(howmuch.index.tolist())

plt.show()
```
